```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from nltk.corpus import stopwords
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
text = "Text analysis is the process of analyzing the textual content for data analysis"
print(text)
```

```python
tokens = word_tokenize(text)
print(tokens)
```

```python
pos_tags = pos_tag(tokens)
print(pos_tags)
```

```python
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in tokens if word.lower() not in stop_words]
print(filtered_words)
```

```python
stemmer = PorterStemmer()
stem_words = [stemmer.stem(word) for word in filtered_words]
print(stem_words)
```

```python
lemmatizer = WordNetLemmatizer()
lem_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print(lem_words)
```

```python
from collections import Counter
tf = Counter(filtered_words)
print(tf)
```
