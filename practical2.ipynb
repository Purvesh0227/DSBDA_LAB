```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data = {
    'name': ['a','b','c','d','e','f','g','h','i','j'],
    'maths': [65, 70, 72, 68, 69, 71, 73, 74, 150, 180],        # outliers: 150, 180
    'science': [85, 88, 90, 87, 89, 91, 86, 88, 300, 350],      # outliers: 300, 350
    'attendance': [60, 62, 65, 63, 64, 66, 61, 62, 200, 220]    # outliers: 200, 220
}

df = pd.DataFrame(data)
df.head()
```

```python
df['maths'] = df['maths'].fillna(df['maths'].mean())
```

```python
df['science'] = df['science'].fillna(df['science'].mean())
df['attendance'] = df['attendance'].fillna(df['attendance'].mean())
df.isnull().sum()
df.columns
```

```python
for col in ['maths', 'science', 'attendance']:
    plt.figure()
    plt.boxplot(df[col])
    plt.title(f"boxplot for {col}")
    plt.ylabel(col)
    plt.show()
```

```python
for col in ['maths', 'science', 'attendance']:
    plt.figure()
    plt.boxplot(df[col])
    plt.title(f"boxplot for {col}")
    plt.ylabel(col)
    plt.show()
```

```python
for col in ['maths', 'science', 'attendance']:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    
    df[col] = df[col].clip(lower, upper)
```

```python
for col in ['maths', 'science', 'attendance']:
    plt.figure()
    plt.boxplot(df[col])
    plt.title(f"boxplot for {col}")
    plt.ylabel(col)
    plt.show()
```
